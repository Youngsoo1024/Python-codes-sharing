```python
import random

suits = ["Clubs", "Diamonds", "Hearts", "Spades"]
faces = ["J", "Q", "K", "A"]
numbered = [str(n) for n in range(2, 11)]

for suit in suits:
    for card in faces+numbered:
        deck.add((card, "of", suit))
card=random.choice(list(deck))
print(card)
deck.remove(card)
print(type(card))
len(deck)
```
