```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the ODE: dy/dt = -ky
def model(y, t, k):
    dydt = -k * y
    return dydt

# Initial condition
y0 = 10

# Time points
t = np.linspace(0, 5, 100)

# Parameters
k = 2

# Solve the ODE
y = odeint(model, y0, t, args=(k,))

# Plot the results
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('Solution of dy/dt = -ky')
plt.grid(True)
plt.show()
```
