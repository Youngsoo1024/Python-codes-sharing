```python
import matplotlib.pyplot as plt
import numpy as np

matrix = {}
filename = "clean 001.txt"
path = "data/"
full_path = path + filename
brief = "clean 001_rev.txt"
with open(full_path, 'r') as file:
   lines = [line.rstrip() for line in file]
   name = lines[0].split(",")
for v in range(1, len(lines)):
   column = lines[v].split(",")
   vg = float(column[3])
   Id = float(column[6])
   R = float(column[5])/float(column[6])
   matrix[v] = [vg]
   
   with open(brief, 'a') as f:
         f.write(f"{vg},{Id},{R}\n")
   plt.scatter(vg, Id, color='blue', s=1)
   plt.ylabel('Id (A)')
   plt.xlabel('V_g (V)')
print(matrix)
#dy = np.diff(matrix)
#dx = np.diff(vg)
#derivative = dy/dx
    
```
