```python
times = ['1:23.43', '2:32.22' , '1:44.50', '1:55.32', '1:43.23']
print(times)
converts =[]
for t in times:
    minutes, rest = t.split(":")
    seconds, hundredths = rest.split(".")
    converts.append((int(minutes)*60*100)+(int(seconds)*100)+int(hundredths))
```

```python
print(converts)
```

```python
times
```

```python
import statistics

#mean value calculation
average = statistics.mean(converts)
average / 100
min_sec, hundreth = str(round(average/100,2)).split(".")
min_sec = int(min_sec)
minutes = min_sec//60
minutes
seconds = min_sec - minutes*60
seconds
minutes, seconds, hundreth
```
