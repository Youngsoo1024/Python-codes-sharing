```python
def event_lookup(event):

    conversions = {
        "Free": "freestyle",
        "Back": "backstroke",
        "Breast": "breaststroke",
        "Fly": "butterfly",
        "IM": "individual medley"
    }
    
    *_, distance, stroke = event.removesuffix(".txt").split("-")
    
    return f"{distance} {conversions[stroke]}"

print(event_lookup("Darius-13-100m-Back.txt"))
```
