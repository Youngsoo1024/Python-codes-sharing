```python

for n, table in enumerate(tables):
    rows = table.find("tr", mode = "all")
    cols = rows[-1].find("td", mode = "all")
    print(f"{n} ->{len(rows)} rows, {len(cols)} cols")
```

```python
table = tables[0]
for row in table.find("tr", mode = "all")[1:]:
    cols = row.find("td", mode = "all")
    event = cols[0].text
    time = cols[1].text
    print(f"\t{event} -> {time}")
print()
```

```python
RECORDS = (0, 2, 4, 5)
COURSES = ("LC Men", "LC Women", "SC Men", "SC Women")

for table, course in zip(RECORDS, COURSES):
    print (f"{course}:")
    # The current value of "table" is used to index into the "Tables"
    # list to grab the world records table you need
    for row in tables[table].find("tr", mode = "all")[1:]:
        cols = row.find("td", mode = "all")
        event = cols[0].text
        time = cols[1].text
        # a tab character formats the output by inserting whitespace before each line of output
        print(f"\t{event} -> {time}")
    # when you ask "print" to display nothing, it completely ignores you and displays a blank line
    # instead (which is a "sort of" nothing)
    print()
    
```
